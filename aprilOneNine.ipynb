```python
from pyspark.sql import SparkSession

# Create Spark session with Hive support
spark = SparkSession.builder \
    .appName("Partitioning and Bucketing Example") \
    .enableHiveSupport() \
    .getOrCreate()

# Sample data
data = [
    (1, "Alice", "USA", 2023),
    (2, "Bob", "India", 2023),
    (3, "Charlie", "USA", 2024),
    (4, "David", "India", 2024),
    (5, "Eve", "UK", 2023),
    (6, "Frank", "UK", 2024)
]

columns = ["id", "name", "country", "year"]
df = spark.createDataFrame(data, columns)
# Write partitioned data to disk
df.write \
    .mode("overwrite") \
    .partitionBy("country") \
    .csv("/output/partitioned_data")

```
